```python
data.columns
```

```python
x = data.iloc[:,1:] # independent 
y = data.iloc[:,0] # dependent 
```

```python
### Feature Importance

from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(x,y)
```

```python
model.feature_importances_
```

```python
# plot graph for feature importance
imp = pd.Series(model.feature_importances_,index=x.columns)
imp.nlargest(8).plot(kind = 'barh')
plt.show()
```

```python
# Present_Price shows the max correlation with selling Price 
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size = 0.2, random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
### hyperparameters

# number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12) ]
# no. of features to consider at every split
max_features = ['auto','sqrt']
# maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5,30, num=6)]
# min no. of samples required to split a node
min_samples_split = [2,5,10,15,100]
# min no. of samples required at each leaf node
min_samples_leaf = [1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV # faster than gridsearchCV
# helps us to find the best parameter
```

```python
# create the random grid
random_grid = { 'n_estimators': n_estimators,
              'max_features':max_features,
              'max_depth':max_depth,
              'min_samples_split':min_samples_split,
              'min_samples_leaf':min_samples_leaf }
print(random_grid)
```

```python
 rf_random = RandomizedSearchCV(estimator=rf, param_distributions = random_grid, scoring = 'neg_mean_squared_error',
                                n_iter = 10, cv = 5, verbose = 2, random_state=42, n_jobs=1)
    # n_jobs = no. cores to be used for faster execution
    # verbose is to display all the fitting done behind with each params
```

```python
rf_random.fit(xtrain,ytrain)
```

```python
rf_random.best_params_
```

```python
pred = rf_random.predict(xtest)
```

```python
sns.distplot(ytest-pred) # if it is forming a normal distribution then it is giving the best result
```

```python
plt.scatter(ytest,pred)
```
